```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
legit =df[df.Class==0]
fraud=df[df.Class==1]
```

```python
legit.value_counts()
```

```python
legit
```

```python
fraud.value_counts()
```

```python
fraud.shape
```

```python
legit.shape
```

```python
legit_sample=legit.sample(n=177)#equal number f 0s and 1s for training and testing so both equal to  177
df_concat=pd.concat ([legit_sample,fraud],axis=0)#concating both the legit and fraud axis
```

```python
df_concat['Class'].value_counts()
```

```python
df_concat.groupby('Class').mean()
```

**Trainig nad testing**

```python
X=df_concat.drop(columns='Class',axis=1)
Y=df_concat['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
#train and test split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
ypred=model.predict(X_test)
```

```python
accuracy_score=accuracy_score(ypred,Y_test)
```

```python
print(accuracy_score)
```

```python
#load the model using the pickle file
import pickle
filename='model.pkl'
pickle.dump(model,open(filename,'wb'))
```
